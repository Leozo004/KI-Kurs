# Bilderkennung

## Einleitung

Bilderkennung ist ein spannendes Thema, mit welchem sich schon seit Jahren in der Informatik beschäftigt wird. Nur durch aktuelle Fortschritte in Big Data und künstlicher Intellegenz, ergeben sich viele neue Möglichkeiten in diesem Themengebiet. Es ist eine Unterkategorie von Computer Vision und KI. Sie befasst sich mit Software zur Erkennung von Objekten, Orten, Personen, Schriften und Aktionen.

## Funktionsweise 

Mit der Bilderkennung werden einzelne Objekte eines Bildes segmentiert. Hierbei werden verschiedene Ressourcen benötigt wie Computer, Bildverarbeitungstechnologien, Kameras und Software auf KI-Basis. So eine Software ahmt Prozesse nach, um Objekte in einem Bild zu erkennen. Mithilfe eines Algorithmuses wird das Bild in Raster und dannach in Pixel unterteilt. Diese Pixel stehen für nummerische Werte. Welche vom Programm zugeordnet werden müssen. Durch Deep-Learning Algorithmen kann so eine Software die Eigenschaften eines Bildes erkennen, diese werden dann mit bestimmten Schwellwertparametern vergleichen. Damit das ganze funktioniert müssen 4 Schritte befolgt werden.

    -Extraktion der Pixelmerkmale
    -Vorbereitung beschrifteter Bilder zum Training
    -Training des Modells
    -Vorhersage eines neuen Bildes(erkennen eines Bildes)

## KI-Methoden zur Computer Vision

### Thresholding
Mit Thresholding kann ein Bild, auf bestimmte (Farb-) Eigenschaften gefiltert werden. Es gibt viele Thresholding-Methoden eine davon ist das Binary-Thresholding, wobei ein Schwellwert definiert wird und ein schwarz/weiß Bild generiert. Wobei Pixel die den zuvor festgelegten Wert überschreiten weiß sind und alle anderen schwarz sind. Diese sind oft Grundlagen für andere Analysen

### Gesichtserkennung

Bei der Gesichtserkenung wird mit "cascades" gearbeitet, diese dienen als eine Art Merkblatt woran sich das Programm orientieren kann. Es gibt viele solcher cascades. Man kann sie auch selbst recht schnell eine anfertigen. Im Bsp-Code wird eine solche cascade genutzt um das frontale Gesicht zu erkennen.

## Beispiele der Nutzung von Bilderkennung

Die Bilderkennung kann in Versicherung, Finanzwesen, E-Commerce, Medizin, Bildung und beim autonohmen Fahren eingesetzt werden.

### Versicherung

Zur Intepretation von Schadensbildern und Berechnung des entstandenen Schadens.

### Finanzwesen

Überweisungen und Abhebungen wären sehr viel einfacher und sicher mit Gesichtserkennung, da Überweisungen nur genemigt werden wenn du sie selber abgibst.

### E-Commerce

Mithilfe eines Bildes kann das gewünschte Produkt gefunden werden.

### Medizin

Mit Bilderkennung, können Krankheiten erkannt werden, indem CT-, Röntgenbilder etc gezeigt werden.
Eine KI kann mit millionen von solchen Bildern gefüttert werden, was sich ein Mensch nicht alles merken könnte.

### Bildung

Bilderkennungssoftware kann eingesetzt werden um Bildungsprozesse zu verbessern, was gerade beim Online-Untericht wichtig werden könnte.

### autonohmes Fahren

Mithilfe von Bilderkennung kann die KI die Straße, Objekte und Witterung erkennen und so ein Fahrzeug wahrscheinlich sicher fahren.

## Code zur Veranschaulichung

Es müssen OpenCV--Headless sowie matplotlib installiert werden. Matplotlib deswegen, weil die Bilder als Diagramm dargestellt werden müssen(in JupyterLab geht es nicht anders).

```python
import cv2
import matplotlib.pyplot as plt

# einelesen des bspBildes
cascade = cv2.CascadeClassifier("/home/jovyan/KI-Kurs/haarcascade_frontalface_default.xml")
image = cv2.imread("/home/jovyan/KI-Kurs/avengers_bild.jpg")
image2 = cv2.imread("/home/jovyan/KI-Kurs/avengers_bild.jpg")

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# cascade Nutzung und zeichnen der Rectangles
faces = cascade.detectMultiScale(image_gray)
for x, y, width, height in faces:
    cv2.rectangle(image, (x, y), (x + width, y + height), color=(0,255,0), thickness=5)

# umwandeln in plt
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Before')
plt.axis('off')
plt.show()
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('After')
plt.axis('off')
plt.show()
```
